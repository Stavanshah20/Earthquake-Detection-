```python
df1 = df.dropna(thresh=df.shape[0]*0.6, axis=1)
```

```python
df1.isnull().sum()
```

```python
from sklearn.impute import KNNImputer
knn = KNNImputer(n_neighbors=3)
knn_df = df1.copy()
knn = knn.fit(knn_df[["Depth"]])
knn_df["Depth"] = knn.transform(knn_df[["Depth"]])

knn_df.isnull().sum()
```

```python
df2 = knn_df.copy()
df2['Place'] = df2['Place'].astype(str)
```

```python
for index, row in df2.iterrows():
    df2.loc[index, 'Country'] = row['Place'].split(" ")[-1]
```

```python
df2
```

```python
df2['Country'].value_counts()
```

```python
df2['Country'].unique()
```

```python
df3_country_other = df2.copy()
df3_country_other['Country'].value_counts().loc[lambda x : x<6]
```

```python
counts = df3_country_other['Country'].value_counts()
counts_df = counts[counts < 6].to_frame().reset_index()
```

```python
counts_df
```

```python
others = []
for i in counts_df['index']:
    others.append(i)
others
```

```python
for i in range(len(df3_country_other)):
    if df3_country_other['Country'][i] in others:
        df3_country_other["Country"][i] = 'others'
```

```python
df3_country_other[df3_country_other['Country'] == "nan"].dropna()
```

```python
df3_country_other = df3_country_other.loc[~df3_country_other['Place'].str.contains('nan')]
```

```python
required_columns = ['Latitude','Longitude','Depth','Mag','MagType','Type','magSource','Country']
all_columns = df3_country_other.columns
not_required_columns = [x for x in all_columns if x not in required_columns]
```

```python
not_required_columns
```

```python
df3_country_other.drop(columns=not_required_columns, inplace=True)
```

```python
dumies = pd.get_dummies(df3_country_other.Country)

dumies.head(10)
```

```python
df4 = pd.concat([df3_country_other,dumies.drop('region',axis='columns')],axis='columns')

df4.head(10)
```

```python
df5 = df4.drop('Country',axis='columns')
df5.head(2)
```

```python
dumies_magsource = pd.get_dummies(df3_country_other.magSource)
dumies_type = pd.get_dummies(df3_country_other.Type)

df_final = pd.concat([df5,dumies_magsource,dumies_type],axis='columns')

df_final_1 = df_final.drop(columns=['Type','magSource'],axis='columns')
df_final_1.head(10)
```

```python
nonnumeric={'MagType'}

from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

for i in nonnumeric:
  df_final_1[i]= label_encoder.fit_transform(df_final_1[i])
df_final_1
```

```python
df_kahaan = df_final_1.copy()
df_kahaan.to_csv("intermediate.csv")
```
